```python
import osmnx as ox
import geopandas as gpd
import folium


gdf = gpd.read_file('DADOS_MS.shp')

# Calcular o centróide para centralizar o mapa
centroid = gdf.geometry.centroid.iloc[0]

m = folium.Map(location=[centroid.y, centroid.x], zoom_start=12)


folium.GeoJson(
    gdf,
    name="geojson",
    style_function=lambda x: {
        'fillColor': 'red',
        'color': 'black',
        'weight': 20,
        'fillOpacity': 1,
    }

).add_to(m)

folium.LayerControl().add_to(m)

m.save('edgewood_map.html')
m
```
